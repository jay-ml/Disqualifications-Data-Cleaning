```python
ndf = pd.DataFrame()
first = True
for index, row in df.iterrows():
    for i in range(1,11):
        n = str(i)
        if row['Penalized School '+n] == '':
            continue
        l = []
        l = [ row['Submission Time'], row['Game Date'], row['Gender'], 
             row['Home'], row['Visitor'], row['Non-Member'], 
             row['Home AD'], row['Away AD'] ]
        l.extend([ row['Penalized School '+n], row['First Name '+n].upper(), row['Last Name '+n].upper(), 
                   row['First Name '+n].upper() + ' ' + row['Last Name '+n].upper(),
                   row['Position '+n], row['Uniform '+n], row['Type '+n] ])
        if row['Type '+n] == 'Yellow Card':
            l.extend([ row['Yellow '+n], row['Yellow Specs '+n] ])
        else:
            l.extend([ row['Red '+n], row['Red Specs '+n] ])
        l.extend([ row['Submitter'].upper(), row['Submitter Team'], row['Submitter Email'], index, i ])
        ldf = pd.DataFrame([l], columns=['Submission Time', 'Game Date', 'Gender', 'Home', 'Visitor',
       'Non-Member', 'Home AD', 'Away AD', 'Penalized School', 'First Name',
       'Last Name', 'Full Name', 'Position', 'Uniform', 'Type', 'Code',
       'Details', 'Submitter', 'Submitter Team', 'Submitter Email', 'Source Index','Game Penalty Index'])
        ndf = pd.concat([ndf, ldf], ignore_index=True)
    print('Game Date: ', row['Game Date'], '| Updated Shape: ', ndf.shape)
print(ndf.head())
```

```python
df = ndf
df = df.loc[df['Penalized School'] != 'Other - for non-member school']
```

```python
def tournamentEligibility(df):
    totals = {}
    reds = {}
    for school in df['Penalized School'].unique():
        curS = df.loc[df['Penalized School'] == school]
        for gen in curS['Gender'].unique():
            cur = curS.loc[curS['Gender'] == gen]
            t = cur.shape[0]
            if 'Red Card' in cur['Type'].unique(): 
                red = cur['Type'].value_counts()['Red Card']
            else:
                red = 0
            totals[(school, gen)] = t
            reds[(school, gen)] = red
    totals_sorted = {k: v for k, v in sorted(totals.items(), reverse=True, key=lambda item: item[1])}
    reds_sorted = {k: v for k, v in sorted(reds.items(), reverse=True, key=lambda item: item[1])}
    print("Totals: ", totals_sorted)
    print("")
    print("Reds: ", reds_sorted)

tournamentEligibility(df)
```

```python
# 5 yellow cards: Player
def findPlayerCards(df):
    yellows = {}
    reds = {}
    totals = {}
    for school in df['Penalized School'].unique():
        curSchool = df.loc[df['Penalized School'] == school]
        for player in curSchool['Full Name'].unique():
            curPlayer = curSchool.loc[curSchool['Full Name'] == player]
            position = curPlayer['Position'].unique()[0]
            totals[(player, school, position)] = 0
            if 'Yellow Card' in curPlayer['Type'].unique():
                yellows[(player, school, position)] = curPlayer['Type'].value_counts()['Yellow Card']
                totals[(player, school, position)] = yellows[(player, school, position)]
            if 'Red Card' in curPlayer['Type'].unique():
                reds[(player, school, position)] = curPlayer['Type'].value_counts()['Red Card']
                totals[(player, school, position)] = totals[(player, school, position)] + reds[(player, school, position)]

    reds_sorted = {k: v for k, v in sorted(reds.items(), reverse=True, key=lambda item: item[1])}
    yellows_sorted = {k: v for k, v in sorted(yellows.items(), reverse=True, key=lambda item: item[1])}
    print("Yellows: ", yellows_sorted)
    print("")
    print("Reds: ", reds_sorted)


findPlayerCards(df)
```

```python
def checkDuplicates(df):
    for gender in df['Gender'].unique():
        gdf = df.loc[df['Gender'] == gender]
        for date in gdf['Game Date'].unique():
            curDate = gdf.loc[df['Game Date'] == date]
            for hs in curDate['Home'].unique():
                curSchool = curDate.loc[curDate['Home'] == hs]
                if len(curSchool['Submitter Email'].unique()) > 1 and len(curSchool['Submitter'].unique()) > 1:
                    print("DUPLICATE SUBMISSION FOUND")
                    print(curSchool)
        print("-----")

checkDuplicates(df)
```
